Increasing disk size on ec2 is as simple as:

from EC2 Dahsboard

click on Volumes
find your volume if you have not explicitly named your volue on ec2 setup identify from the 'Attachment Information' column.
Click Actions and modify volume.
Add/change volume size and click modify.
In the Description tab (bottom of screen) check the state, this will update as the tasks process. I could continue to use my instance as the disk was enlarging. I reset my instance (don't think you need to) as when you initially ssh after reboot you see the disk space available.


To create an S3 bucket

start creating folders and uploadind data.

I've creates a new AIM role of futures

Click Actions --> Instance Settings --> Attach/Replace IAM Role
select futures in AIM role dropdown and clieck Apply.

To access S3 from python:

```python
from boto.s3.connection import S3Connection
 
aws_connection = S3Connection()
bucket = aws_connection.get_bucket('futures-deep-learning')
for file_key in bucket.list():
    print file_key.name

myfile = bucket.get_key('/dstl/three_band/6180_4_4.tif')
myfile.get_contents_to_filename('/home/ubuntu/data/6180_4_4.tif')
```
